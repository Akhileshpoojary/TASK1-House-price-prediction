```python
# Assuming 'price' is your target variable
labels = data['price']

# Convert 'date' column
conv_yr_built = [1 if year == 2014 else 0 for year in data.yr_built]
data['yr_built'] = conv_yr_built

# Drop 'price' column and use 'yr_built' instead
train1 = data.drop(['price'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(train1 , labels , test_size = 0.10,random_state =2)
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python

from sklearn.ensemble import GradientBoostingRegressor

# Assuming you have defined x_train and y_train
clf = GradientBoostingRegressor(loss='squared_error')  # Change 'ls' to 'squared_error'
clf.fit(x_train, y_train)
```

```python
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error

# Assuming you have x_train, y_train, x_test, and y_test defined

# Define hyperparameters
params = {
    'n_estimators': 100,
    'learning_rate': 0.1,
    'max_depth': 3,
    'min_samples_split': 2,
    'min_samples_leaf': 1,
    'subsample': 1.0,
    'loss': 'squared_error'
}

# Create a GradientBoostingRegressor with the specified hyperparameters
clf = GradientBoostingRegressor(**params)

# Fit the model with training data
clf.fit(x_train, y_train)

# Now you can use staged_predict
t_sc = np.zeros((params['n_estimators']), dtype=np.float64)
for i, y_pred in enumerate(clf.staged_predict(x_test)):
    t_sc[i] = mean_squared_error(y_test, y_pred)  # Use appropriate loss function
testsc = np.arange((params['n_estimators']))+1
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(testsc,clf.train_score_,'b-',label= 'Set dev train')
plt.plot(testsc,t_sc,'r-',label = 'set dev test')
```
